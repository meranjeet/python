Processing Array

```python
import numpy as np

import matplotlib.pyplot as plt
```

```python
points = np.arange(-5,5,0.01)
points
```

```python
dx,dy=np.meshgrid(points,points)
dx
```

```python
dy
```

```python
z=np.sin(dx)+np.sin(dy)
z
```

```python
plt.imshow(z)
plt.colorbar()
plt.title('Plot of Sin(x)+Sin(y)')
```

```python
A=np.array([1,2,3,4])

B=np.array([100,200,300,400])

condition =np.array([True,True,False,False])

answer=[(A_val if cond else B_val) for A_val,cond,B_val in zip(A,condition,B)]

answer
```
